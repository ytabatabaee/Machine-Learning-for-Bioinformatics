```
import torch
import numpy as np
from torch.autograd import Variable
```

## Tensors

```
a = np.zeros(shape=(3, 4))
print(a.dtype, type(a))
print(a)
```

```
b = torch.zeros(3, 4, dtype=torch.int32)
print(b.dtype, type(b))
print(b)
```

```
# numpy to tensor
c = torch.tensor(a)
print(c.dtype, type(c))
print(c)
```

```
# tensor to numpy
b.numpy()
```

## Operations

```
print(torch.add(c, b))
print(b + c)
```

```
res = torch.empty(3,4) # empty allocates uninitialized memory
torch.add(b, c, out=res)
res
```

```
# matrix multiplication with matmul or @

a = torch.randn(3,4)
b = torch.randn(3,)
c = torch.matmul(b,a)
c, b@a
```

```
# elementwise multiplication with *

b = torch.randn(3,4)
a*b
```

## Autograd

```
x = torch.ones(2, 2, requires_grad=True)
print(x)
print(x.grad)
```

```
y = x + 2
y
```

```
y.grad_fn
```

```
z = y.mean()
z
```

```
z.backward()
```

```
print(x.grad)
```

```
x = x - 0.001 * x.grad
```

```
x
```

```
# doesn't give error in inplace gradiant computation
X = Variable(torch.ones(2, 2), requires_grad=True)
X
```
