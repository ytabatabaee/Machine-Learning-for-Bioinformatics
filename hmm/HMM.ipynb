```
from hidden_markov import hmm
# h: healthy , i: ill
states = ('h', 'i')
# S: sleep, E: eat, A: Exercise
possible_observation = ('S', 'E', 'A')

# Initial Variables
start_probability = np.matrix('0.4 0.6')
transition_probability = np.matrix('0.5 0.5 ; 0.5 0.5')
emission_probability = np.matrix('0.3 0.3 0.4; 0.4 0.3 0.3')
```

```
# Baum-welch Algorithm with hidden_markov library

# Initialize the model
model = hmm(states, possible_observation, start_probability, transition_probability, emission_probability)
observations = [('E', 'A', 'E', 'S', 'S', 'E', 'A', 'E', 'S', 'S', 'A', 'E')]
quantities_observations = [1]
```

```
num_iter = 100
e, t, s = model.train_hmm(observations, num_iter, quantities_observations)
```
